# Data Cleaning

```python
def cleanCorruptedFile(CorruptedFile, CleanFileName):
    with open(CorruptedFile, "r") as file:
        text = file.readlines()
    
    for i in range(len(text)-1, -1, -1):
        if i != len(text)-1:
            text[i] = text[i][:-1]
        if text[i][:1] == '*':
            text[i] = text[i][::-1]
        if text[i][:2] == '!!':
            text.pop(i)
        else:
            text[i] = text[i][4:-4]
    
    with open(CleanFileName, "w") as file:
        for row in text:
            file.write(row + '\n')
```

# Data Visualization

```python
def VisualizeCategories(data):
    with open(data, "r") as file:
        text = file.readlines()
        
    category_count = {'OPERATIONAL': 0, 'TACTICAL': 0, 'STRATEGIC': 0}    
    for i in range(0, len(text)):
        text[i] = text[i][:-1].split()
        category_count[text[i][7].upper()] += 1
        
    fig = plt.figure()
    ax = fig.add_axes([0,0,1,1])
    ax.axis('equal')
    categories = list(category_count.keys())
    counts = list(category_count.values())
    ax.pie(counts, labels = categories, autopct='%1.1f%%', explode=[0.1, 0, 0], 
           shadow=True, startangle=210, colors=['darkorange', 'forestgreen', 'royalblue'])
    
    return plt.show()
```

```python
VisualizeCategories('SC2TGT_ListClean')
```

```python
def VisualizeRaces(data):
    with open(data, "r") as file:
        text = file.readlines()
        
    race_count = {'ZERG': 0, 'PROTOSS': 0}    
    for i in range(0, len(text)):
        text[i] = text[i][:-1].split()
        race_count[text[i][1].upper()] += 1
        
    fig = plt.figure()
    ax = fig.add_axes([0,0,1,1])
    ax.axis('equal')
    races = list(race_count.keys())
    counts = list(race_count.values())
    ax.pie(counts, labels = races, autopct='%1.1f%%', explode=[0, 0], 
           shadow=True, startangle=275, colors=['darkorange','royalblue'])
    
    return plt.show()
```

```python
def Target_type_counts(designation):
    dict_counts = {'Resource': 0, 'Depot': 0, 'Barracks': 0, 'System': 0, 'Unit': 0, 'Obstacle': 0, 'Fortress': 0}
    for i in text:
        if i[0][:2] == designation:
            dict_counts[i[9]] += 1        
    print('TARGET TYPE COUNTS' + designation +  ' DESIGNATED TARGETS\n    ', 'Resources:', dict_counts['Resource'], 
         'Depot:', dict_counts['Depot'], 'Barracks:', dict_counts['Barracks'], 'System:', dict_counts['System'],
         'Unit:', dict_counts['Unit'], 'Obstacle:', dict_counts['Obstacle'], 'Fortress:', dict_counts['Fortress'])
    
def Target_priority_counts(designation):
    dict_counts = {'Urgent': 0, 'High': 0, 'Routine': 0}
    for i in text:
        if i[0][:2] == designation:
            dict_counts[i[2]] += 1        
    print('TARGET PRIORITY COUNTS' + designation +  ' DESIGNATED TARGETS\n    ', 'Urgent:', dict_counts['Urgent'], 
         'High:', dict_counts['High'], 'Routine:', dict_counts['Routine'])
```

# Data Modelling

```python
@dataclass
class  SC2Target:
    Target_Designation: str
    Target_Race: str
    Target_Priority: str
    Target_Shape: str
    Target_Length: int
    Target_Width: int
    Number_of_Target_Components: int
    Target_Category: str
    Target_Surveillance: str
    Target_Type: str
    Target_Density: str
    Target_Mobility: str
    Target_Protection: str
    Target_X_Coordinate: int
    Target_Y_Coordinate: int
    Target_Observer: str
    Target_Discovered_Game_Time: int
    Target_Health: int
    Target_Status: str
    Target_Point_Value: int
        
    def display(self):
        print('IDENTIFICATION\n    TGD ID:', self.Target_Designation, 'Race:', self.Target_Race, 'Priority:', self.Target_Priority)
        print('CONFIGURATION\n    Shape:', self.Target_Shape, 'Length:', self.Target_Length, 'Width:', self.Target_Width, 'Aim Points:', self.Number_of_Target_Components)
        print('CATEGORY\n    Category:', self.Target_Category, 'Surveillance:', self.Target_Surveillance, 'Type:', self.Target_Type)
        print('TARGETABILITY\n    Density:', self.Target_Density, 'Mobility:', self.Target_Mobility, 'Protection:', self.Target_Protection)
        print('LOCATION\n    X:', self.Target_X_Coordinate, 'Y:', self.Target_Y_Coordinate)
        print('OBSERVER\n    Located By:', self.Target_Observer, 'At Time:', self.Target_Discovered_Game_Time)
        print('VALUE\n    Health:', self.Target_Health, 'Status:', self.Target_Status, 'Value:', self.Target_Point_Value)

def buildAllTargets(data):
    data_arr = [None] *1000
    
    with open(data, "r") as file:
        text = file.readlines()

    for i in range(0, len(text)):
        text[i] = text[i][:-1].split()
        text[i][4] = int(text[i][4])
        text[i][5] = int(text[i][5])
        text[i][6] = int(text[i][6])
        text[i][13] = int(text[i][13])
        text[i][14] = int(text[i][14])
        text[i][16] = int(text[i][16])
        text[i][17] = int(text[i][17])
        text[i][19] = int(text[i][19])
        
    for i in range(0, len(text)):
        data_arr[i] =  SC2Target(text[i][0], text[i][1], text[i][2], text[i][3], text[i][4], text[i][5], text[i][6], text[i][7], text[i][8], text[i][9], text[i][10],
             text[i][11], text[i][12], text[i][13], text[i][14], text[i][15], text[i][16], text[i][17], text[i][18], text[i][19])
        
    return data_arr
```

```python
SC2Target_list = buildAllTargets('SC2TGT_ListClean')
```

```python
def showTarget(designation, SC2Target_list = buildAllTargets('SC2TGT_ListClean')):
    no_exist = True
    for i in SC2Target_list:
        if designation == i.Target_Designation:
            no_exist = False
            i.display()
    if no_exist:
        print('Target Does Not Exist')
```

```python
def showTgtPriorities(priority, SC2Target_list = buildAllTargets('SC2TGT_ListClean')):
    dict_count = {'Zerg': 0, 'Protoss': 0}    
    for i in SC2Target_list:
        if i.Target_Priority == priority:
            if i.Target_Race == 'Zerg':
                dict_count['Zerg'] += 1
            else:
                dict_count['Protoss'] += 1
                
    
    print('Zerg Priority Routine\n    Target Count is', dict_count['Zerg'], 
         '\n\nProtoss Priority Routine\n    Target Count is', dict_count['Protoss'])
    
showTgtPriorities('Routine')
```

# Data Sampling

```python
def randomTgtSampler(size, SC2Target_list = buildAllTargets('SC2TGT_ListClean')):
    new_list = [SC2Target_list[random.randint(0, len(SC2Target_list)-1)] for i in range(0, size)]
    dict_count = {'Zerg': 0, 'Protoss': 0} 
    zerg_tot = 0; zerg_min = 9999999; zerg_max = 0
    protoss_tot = 0; protoss_min = 9999999; protoss_max = 0
    for i in new_list:
        if i.Target_Race == 'Zerg':
            if i.Target_Point_Value > zerg_max:
                zerg_max = i.Target_Point_Value
            if i.Target_Point_Value < zerg_min:
                zerg_min = i.Target_Point_Value
            dict_count['Zerg'] += 1
            zerg_tot += i.Target_Point_Value
        else:
            if i.Target_Point_Value > protoss_max:
                protoss_max = i.Target_Point_Value
            if i.Target_Point_Value < protoss_min:
                protoss_min = i.Target_Point_Value
            dict_count['Protoss'] += 1
            protoss_tot += i.Target_Point_Value
            
    zerg_mean = round((zerg_tot / dict_count['Zerg']), 2)
    zerg_range = zerg_max - zerg_min
    protoss_mean = round((protoss_tot / dict_count['Protoss']), 2)
    protoss_range = protoss_max - protoss_min
    
    print('Sample Size:', size, '\nRACE: Zerg:', dict_count['Zerg'], 'Protoss:',  dict_count['Protoss'])
    
    print('Zerg Total Target Point Value:', zerg_tot,
            '\nZerg Minimum Target Point Value:', zerg_min,
            '\nZerg Maximum Target Point Value:', zerg_max,
            '\nZerg Mean Target Point Value:',  zerg_mean,
            '\nZerg Target Point Value Range:', zerg_range,
          
              '\n\nProtoss Total Target Point Value:', protoss_tot,
            '\nProtoss Minimum Target Point Value:', protoss_min,
            '\nProtoss Maximum Target Point Value:', protoss_max,
            '\nProtoss Mean Target Point Value:',  protoss_mean,
            '\nProtoss Target Point Value Range:', protoss_range)  
```

```python
randomTgtSampler(10)  
```
