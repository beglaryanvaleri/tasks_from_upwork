```python
import pandas as pd

with open('New13.txt', "r", encoding='UTF-8') as file:
        text = file.readlines()
        
df = pd.DataFrame(columns = ['Date', 'Subject', 'Folder', 'https'])
n = 0
dict1 = {'Date': '', 'Subject': '', 'Folder': '', 'https': ''}
for i in text:
    if i.split(":", 1)[0][:1] == '\n':
        df = df.append(dict1, ignore_index=True)
        dict1 = {'Date': '', 'Subject': '', 'Folder': '', 'https': ''}
    
    if  'Date' in i.split(":", 1)[0]:
        dict1['Date'] = i.split(":", 1)[1][:-1]
        
    if  'Subject' in i.split(":", 1)[0]:
        dict1['Subject'] = i.split(":", 1)[1][:-1]
        
    if  'Folder' in i.split(":", 1)[0]:
        dict1['Folder'] = i.split(":", 1)[1][:-1]
        
    if  'tps' in i.split(":", 1)[0]:
        dict1['https'] += i.split(":", 1)[1][:-1]
        
        
```
